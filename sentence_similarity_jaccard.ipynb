```python
def jacaard_similarity(listA, listB):
    s1 = set(listA)
    s2 = set(listB)
    
    return float(len(s1.intersection(s2)) / len(s1.union(s2)))
```

```python
# 형태소 분리
# df_old['pos_split'] = df_old['발화데이터'].apply(tagger.pos)

def pos_split(df:pd.DataFrame, col='speech'):
    i = -1
    for i in range(df.shape[0]//1000):      # 넘 많아~~
        df.loc[i*1000:(i+1)*1000, 'pos_split'] = df.loc[i*1000:(i+1)*1000, col].astype(str).apply(tagger.pos)
        print(f'{i*1000} ~ {(i+1)*1000} 완')
    df.loc[(i+1)*1000:, 'pos_split'] = df.loc[(i+1)*1000:, col].apply(tagger.pos)
```
